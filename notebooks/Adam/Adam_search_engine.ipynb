### Download image

Download googleimagesdownload package in python (https://google-images-download.readthedocs.io/en/latest/arguments.html)

Put googleimagesdownload -cf config.json in terminal.

### Load downloaded image

```python
import os
import numpy as np
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt 
```

```python
images = []
labels = []
download_folder = os.listdir('downloads')
for folder in download_folder:
    subfolder_name = 'downloads' + '/' + folder
    for image_name in os.listdir(subfolder_name):
        image = mpimg.imread(subfolder_name + '/' + image_name) #depends whether we want to open images by mping
        label = image_name.split(' ', 1)[0]
        
        images.append(image)
        labels.append(label)
```

```python
plt.imshow(images[0]) 
print(labels[0])
```
