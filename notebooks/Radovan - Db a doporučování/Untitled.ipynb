```python
import pandas as pd
import numpy as np
datafile = "../../data/full_format_recipes_parsed_ingredients_final.json"
```

```python
df = pd.read_json(datafile)
df.head()
```

```python
cat, catn = np.unique(np.concatenate(df.loc[:,'categories'].values), return_counts=True)
i = np.argsort(-catn)
cat = cat[i]
catn = catn[i]
print(np.stack([cat, catn]).transpose()[0:20,:])
```

```python
ing, ingn = np.unique(np.concatenate(df.loc[:,'ingredients_preformatted'].values), return_counts=True)
i = np.argsort(-ingn)
ing = ing[i]
ingn = ingn[i]
print(np.stack([ing, ingn]).transpose()[0:200,:])
```
